```python
#-------------------------------Product Classification--------------------------------------------------------------
```

```python
def Classification(x):
    if x<=300:
        return 1
    elif 300<x<=1000:
        return 2
    elif 1000<=x<5000:
        return 3
    else:
        return 4

# Category 1 stands for Low-end Shoes; 2 for Median-end Shoes; 3 for High-end Shoes; 4 Represents for those extremely expensive shoes or orders with potential mistakes.
```

```python
df['category'] = df.unit_price.map(Classification)
df.head()
```

```python
df.category.value_counts()
```

```python
df1=df[df['category']==1]
df2=df[df['category']==2]
df4=df[df['category']==4]
```
